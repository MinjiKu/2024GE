```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 정의
x = { 
         #단위는 billion USD
    'Korea':{'steel':65.85, 'semi':19.8, 'car':188.856},
    'USA': {'steel':4.14, 'semi':80.5, 'car':768.0},
    'China': {'steel':421.18, 'semi':193.6, 'car':772},
    'Japan':{'steel':58.92, 'semi':51.9, 'car':476.784},
    'Germany':{'steel':44.50, 'semi':41.2, 'car': 614.313}
}
```

```python
# 산업 목록
industries = list(next(iter(x.values())).keys())  # 'steel', 'semi', 'car'
countries = list(x.keys())

# 색상 정의
colors = ['#4B4B4B', '#bb0a1e', '#ff7f0e', '#1f77b4', '#2ca02c']

# 산업별로 그래프 그리기
for industry in industries:
    fig, ax = plt.subplots(figsize=(10, 6))

    # 데이터를 준비합니다.
    values = [x[country][industry] for country in countries]
    bars = ax.bar(countries, values, color=colors)

    # 그래프 설정
    ax.set_ylabel('Export Volume (Billion USD)')
    ax.set_xlabel('Country')
    ax.set_title(f'Export Volume for {industry.capitalize()} Industry')
    
    # 값 레이블 추가
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')  # va: vertical alignment

    # 레이아웃 조정
    plt.tight_layout()
    plt.show()
```
